```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
camera =cv2.VideoCapture(0)
```

```python
plt.imshow(img)
plt.show()
```

```python
facedetector=cv2.CascadeClassifier('facetmp.xml')
```

```python
camera = cv2.VideoCapture(0)
name= input("enter the name of person")
pics=[]
cunt=0
while True:
    b,img=camera.read()
    
    if b==False:
        continue
    #detect face
    faces=facedetector.detectMultiScale(img,1.3,5)
    #no facesc detected
    if(len(faces)==0):
        continue
    
    #draw a bounding box
    face=faces[0]
    
    green=(255,0,0)
    
    x,y,w,h=faces[0]
    cv2.rectangle(img,(x,y),(x+w,y+h),green,5)
    cropped_faces=img[y:y+h,x:x+w]
    cv2.imshow("cropped faces",cropped_faces)
    cunt=cunt+1
    if cunt%10==0:
        print("saving_pic",cunt/10)
        cropped_faces = cv2.resize(cropped_faces,(100,100))
        pics.append(cropped_faces)
    
    if cunt/10==20:
        break
    
    
    # Show the New Image
    cv2.imshow("Title",img)
    #Add some delay 1 ms between 2 frames
    cv2.waitKey(1)
    
camera.release()
cv2.destroyAllWindows()

pics = np.array(pics)
print(pics.shape)

np.save(name+".npy",pics)
print("Images saved to disk")

```
